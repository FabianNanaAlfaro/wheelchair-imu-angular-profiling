# Cleaned iSen summary workflow

Cleaned public-facing notebook for the wheelchair IMU angular profiling workflow.


```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
def summarize_angle_series(y):
    y = np.asarray(y, dtype=float)
    return {
        'max_deg': np.nanmax(y),
        'min_deg': np.nanmin(y),
        'range_deg': np.nanmax(y) - np.nanmin(y),
        'sd_deg': np.nanstd(y, ddof=1),
        'mean_deg': np.nanmean(y),
    }
```

```python
# Load processed angle CSV files and summarize them.
input_folder = Path('../outputs/python_exploration')
rows = []
for path in sorted(input_folder.glob('*.csv')):
    df = pd.read_csv(path)
    col = 'angle_deg' if 'angle_deg' in df.columns else df.columns[-1]
    row = summarize_angle_series(df[col])
    row['file'] = path.name
    rows.append(row)
summary = pd.DataFrame(rows)
summary
```
